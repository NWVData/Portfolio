```python
## Import Libraries ##

import pandas as pd
import numpy as np
import datetime as dt
from matplotlib import pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
## Import Movies File ##

movies = pd.read_csv('top10kmovies.csv')
movies.head()
```

```python
## Table Info ##
movies.info()
```

```python
## Drop first and last columns ##
movies = movies.drop(['Unnamed: 0','tagline'], axis = 1)
movies.head()
```

```python
## Drop Overview Column ##
movies = movies.drop('overview', axis = 1)
movies.head()
```

```python
## Change object dtyples to string ##
for col in movies.columns:
  if movies[col].dtype == 'object':
    movies[col] = movies[col].astype('string')
```

```python
## Take First Genre from Genres column to make Main Genre Column ##
movies['main_genre'] = movies['genres'].str.split(',').str.get(0)
movies.insert(3,'main_genre',movies.pop('main_genre'))
movies.head()
```

```python
## Format Main Genre Column ##
movies['main_genre'] = movies['main_genre'].str.strip("[]").str.replace("'", "")
movies.head()
```

```python
## Drop Genres ##
movies = movies.drop('genres',axis=1)
movies.head()
```

```python
## Take First Prod. Company from Prod. Company column to make Main Company Column ##
movies['main_prod_company'] = movies['production_companies'].str.split(',').str.get(0)
movies.insert(9,'main_prod_company',movies.pop('main_prod_company'))
movies.head()
```

```python
## Format Prod. Company Column ##
movies['main_prod_company'] = movies['main_prod_company'].str.strip("[]").str.replace("'", "")
movies.head()
```

```python
## Drop Prod. Companies Column ##
movies = movies.drop('production_companies', axis=1)
movies.head()
```

```python
## Convert Runtime to Hours and Mins ##
movies['runtime_formatted'] = pd.to_timedelta(movies['runtime'], unit='s')
movies['runtime_formatted'] = movies['runtime_formatted'].astype(str).str[7:]
movies.head()
```

```python
## Drop Runtime ##
movies = movies.drop('runtime', axis=1)
movies.head()
```

```python
## Scale Revenue and Budget to a Smaller Number ##
scaling_factor = 1_000_000

movies['scaled_revenue'] = round((movies['revenue'] / scaling_factor),2)
movies.insert(10,'scaled_revenue',movies.pop('scaled_revenue'))
movies['scaled_budget'] = round((movies['budget'] / scaling_factor),2)
movies.insert(8,'scaled_budget',movies.pop('scaled_budget'))
movies.head()
```

```python
## Drop Budget and Revenue Columns ##
movies = movies.drop(['budget','revenue'],axis=1)
movies.head()
```

```python
## Find Duplicates ##
movies.duplicated().sum()
```

```python
## Table Descriptions ##
movies.describe()
```

```python
## Table Shape ##
movies.shape
```

```python
## Final Table Review, Sorted by Revenue ##
movies.sort_values(by='scaled_revenue',ascending=False)
```

```python
## Counts of Genres ##
topgenres = movies.main_genre.value_counts().nlargest(10).reset_index()
topgenres.columns = ['Genre','Count']
print(topgenres)
```

```python
## Countplot of Genres ##

plt.figure(figsize=(10,6))
topgenresplot = sns.barplot(topgenres,x='Genre',y='Count', palette = 'colorblind')
topgenresplot.set(title='Top 10 Genres by Count',xlabel='Genre',ylabel='Genre Count')
plt.axhline(y=np.mean(topgenres['Count']), linestyle='--', color='red', linewidth=3, label='Avg. Count')
plt.tick_params(axis='x', labelsize=12,labelrotation=30,labelcolor='navy')
plt.legend()
sns.despine()
sns.set(style='white')
for labels in topgenresplot.containers:
  topgenresplot.bar_label(labels)
plt.show()
```

```python
## What Movies Had the Most Revenue? ##
movierev = movies.groupby('title').agg({'scaled_revenue':'sum'}).reset_index()

plt.figure(figsize=(5,8))
movierevplot = sns.barplot(movierev.nlargest(5,'scaled_revenue'),x='scaled_revenue',y='title', palette = 'crest')
movierevplot.set(title='Top 10 Movies by Revenue (Scaled)',xlabel='Title',ylabel='Scaled Revenue')
sns.despine()
sns.set(style='white')
for labels in movierevplot.containers:
  movierevplot.bar_label(labels)
plt.show()
```

```python
## What Production Companies Make the Highest Rated Movies? - Pie ##

## Get Top 5 Ratings where there are more than 25k votes ##
movievotes = movies.vote_count >= 25000
prodrating = movies[movievotes].groupby(by='main_prod_company').agg({'vote_average':'mean'})\
.sort_values(by='vote_average', ascending=False).reset_index()

## Get Variable for Chart ##
prodrating5 = prodrating.head(5)

## Make Chart ##
plt.figure(figsize=(8,8))
sns.set_palette('pastel')
plt.pie(prodrating5['vote_average'],labels=prodrating5['main_prod_company'],\
        autopct='%1.1f%%',startangle=60)
sns.set('poster')
plt.title('Rating by Top 5 Production Companies > 25,000 Votes')
plt.show()
```

```python
## Most Profiting Genres - Barplot ##

## Make Profit Column ##
movies['profit'] = movies['scaled_revenue'] - movies['scaled_budget']

## Make variable for plot ##
genrepro = movies.groupby(by='main_genre').agg({'profit':'sum'}).sort_values(by='profit',ascending=False).reset_index()

## Make plot ##
plt.figure(figsize=(12,6))
genprofit = sns.barplot(data=genrepro.nlargest(5,'profit'),x='main_genre',y='profit')
genprofit.set(title='Most Profiting Genres',xlabel='Genre',ylabel='Profit')
for label in genprofit.containers:
  genprofit.bar_label(label)
plt.show()
```
